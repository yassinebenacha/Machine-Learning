```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data"
data = pd.read_csv(url, header=None)
print('columns in dataset:')
print(data.columns)
```

```python
print("Statistical information of the dataset:")
print(data.describe())
```

```python
# describe the ship of the data site
print("Shape of the dataset:", data.shape)
```

```python
#5-Display the number of missing values in each column
print("Number of missing values in each column:")
print(data.isnull().sum())
```

```python
import numpy as np
import pandas as pd  # Ensure you import pandas if you're working with DataFrames

# Assuming 'data' is already defined as a DataFrame
data = data.replace('?', np.nan)  # Replace '?' with NaN

print('number of instances=%d' % (data.shape[0]))  # Number of rows
print('number of attributes=%d' % (data.shape[1]))  # Number of columns

print('number of missing values=')
for col in data.columns:
    # Use isna() to count NaN values
    missing_count = data[col].isna().sum()  
    print('\t%s:%d' % (col, missing_count))
```

```python
# Choose the specific column to fill missing values
column_names = [
    "Sample code number", "Clump Thickness", "Uniformity of Cell Size",
    "Uniformity of Cell Shape", "Marginal Adhesion", "Single Epithelial Cell Size",
    "Bare Nuclei", "Bland Chromatin", "Normal Nucleoli", "Mitoses", "Class"
]

# Display the values before filling missing values
print("Values before filling missing values:")
print(data[column_name].head(10))  # Display first 10 values for brevity

# Calculate the median of the column
median_value = data[column_name].median()

# Fill missing values with the median
data[column_name].fillna(median_value, inplace=True)

# Display the values after filling missing values
print("\nValues after filling missing values:")
print(data[column_name].head(10))  # Display first 10 values for brevity
```
